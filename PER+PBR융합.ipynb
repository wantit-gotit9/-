```python
import pandas as pd
import numpy as np
```

```python
#XLSX 파일(KRX 코스피) 불러오기
df=pd.read_excel('/content/data_PER+PBR.xlsx', index_col=0)
df
```

```python
#보기 편하게 행열 교체
df=df.transpose()
df
```

```python
#'-'스트링을 NaN으로 변환
def isnumber(x):
    try:
        float(x)
        return True
    except:
        return False

df = df[df.applymap(isnumber)]
df
```

```python
PER = df.iloc[-5]
PER
```

```python
PBR = df.iloc[-3]
PBR
```

```python
df = pd.DataFrame(df, index='PER PBR'.split())
df = df.transpose()
df
```

```python
df['PER rank']=df['PER'].rank()
df['PBR rank']=df['PBR'].rank()
df['Total rank']=df['PER rank']+df['PBR rank']
df
```

```python
df=df.dropna()
df
```

```python
df=df.sort_values(by='Total rank')
df[:10].index.tolist()
```
